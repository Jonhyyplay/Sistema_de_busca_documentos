```python
#Listagem de documentos que serão buscados
DOCUMENT1 = {
    "Título": "Operando o sistema de controle climático",
    "conteúdo": "Seu GoogleCar possui um sistema de controle climático que permite ajustar a temperatura e o fluxo de ar no carro. Para operar o sistema de controle climático, use os botões e botões localizados no console central. Temperatura: o botão de temperatura controla a temperatura dentro do carro. Gire o botão no sentido horário para aumentar a temperatura ou no sentido anti -horário para diminuir a temperatura. Fluxo de ar: o botão de fluxo de ar controla a quantidade de fluxo de ar dentro do carro. Gire o botão no sentido horário para aumentar o Airf… D: O botão de velocidade do ventilador controla a velocidade do ventilador. Gire o botão no sentido horário para aumentar a velocidade do ventilador ou no sentido anti -horário para diminuir a velocidade do ventilador. Modo: o botão Modo permite selecionar o modo desejado. Os modos disponíveis são: Auto: o carro ajustará automaticamente a temperatura e o fluxo de ar para manter um nível confortável. Legal: o carro soprará ar fresco no carro.Calor: O carro soprará ar quente no carro. Descongelar: o carro soprará ar quente no para -brisa para descongelar."}

DOCUMENT2 = {
    "Título": "Touchscreen",
    "conteúdo": "Seu GoogleCar possui uma tela sensível ao toque grande que fornece acesso a uma variedade de recursos, incluindo navegação, entretenimento e controle climático. Para usar a tela sensível ao toque, basta tocar no ícone desejado. Por exemplo, Você pode tocar no ícone \ navegação \" para obter instruções para o seu destino ou tocar no ícone \ music \ para tocar suas músicas favoritas."}

DOCUMENT3 = {
    "Título": "Mudança de Marchas",
    "conteúdo": "Seu googlecar tem uma transmissão automática. Para mudar de engrenagem, basta mover a alavanca de mudança para a posição desejada. Park: Esta posição é usada quando você está estacionado. As rodas estão trancadas e o carro não pode se mover. Reverso: esta posição é usada para fazer backup. Neutro: esta posição é usada quando você é parado em uma luz ou no tráfego. O carro não está em marcha e não se moverá a menos que você pressione o pedal do acelerador. Drive: esta posição é usada para dirigir para a frente. Baixo: Esta posição é usada para dirigir em neve ou outras condições escorregadias."}

documents = [DOCUMENT1, DOCUMENT2, DOCUMENT3]
```

```python
df = pd.DataFrame(documents)
df.columns = ["Titulo", "Conteudo"]
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Titulo"], row["Conteudo"]), axis=1)
df
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
 embedding_da_consulta = genai.embed_content(model=model,
                                 content=consulta,
                                 task_type="RETRIEVAL_QUERY")["embedding"]
 produtos_escalares = np.dot(np.stack(df["Embeddings"]), embedding_da_consulta)

 indice = np.argmax(produtos_escalares)
 return df.iloc[indice] ["Conteudo"]
```

Teste 1

```python
consulta = "Como faço para trocar a marcha no carro da Google?"

trecho = gerar_e_buscar_consulta(consulta, df, model)
print(trecho)
```

```python
consulta = "Como funciona o Touchscreen do carro da Google?"

trecho = gerar_e_buscar_consulta(consulta, df, model)
print(trecho)
```
